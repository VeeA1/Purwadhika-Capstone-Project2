### Dataset Explanation
Dataset “Supermarket Customer” berisi deskripsi detail data pelanggan supermarket ideal suatu perusahaan. Kumpulan data ini berisi 29 kolom yang merinci kebiasaan dan demografi pelanggan. Dan masing-masing dari 29 kolom tersebut dibagi menjadi 4 kategori, berikut detailnya :

**People :**

ID : Customer’s unique identifier

Year_Birth : Customer’s birth year

Education : Customer’s education level

Marital_Status : Customer’s marital status

Income : Customer’s yearly household income

Kidhome : Number of children in customer’s household

Teenhome : Number of teenagers in customer’s household

Dt_Customer : Date of customer’s enrollment with the company

Recency : Number of days since customer’s last purchase

Complain : 1 if the customer complained in the last 2 years, 0 otherwise

**Products**

MntWines : Amount spent on wine in last 2 years

MntFruits : Amount spent on fruits in last 2 years

MntMeatProducts : Amount spent on meat in last 2 years

MntFishProducts : Amount spent on fish in last 2 years

MntSweetProducts : Amount spent on sweets in last 2 years

MntGoldProds : Amount spent on gold in last 2 years

**Promotion**

NumDealsPurchases : Number of purchases made with a discount

AcceptedCmp1 : 1 if the customer accepted the offer in the 1st campaign, 0 otherwise

AcceptedCmp2 : 1 if the customer accepted the offer in the 2nd campaign, 0 otherwise

AcceptedCmp3 : 1 if the customer accepted the offer in the 3rd campaign, 0 otherwise

AcceptedCmp4 : 1 if the customer accepted the offer in the 4th campaign, 0 otherwise

AcceptedCmp5 : 1 if the customer accepted the offer in the 5th campaign, 0 otherwise

Response : 1 if the customer accepted the offer in the last campaign, 0 otherwise

**Place**

NumWebPurchases : Number of purchases made through the company’s website

NumCatalogPurchases : Number of purchases made using a catalog

NumStorePurchases : Number of purchases made directly in stores

NumWebVisitsMonth : Number of visits to the company’s website in the last month

```python
corr_Acc = df_marketCL[['Income','AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4',
                           'AcceptedCmp5', 'AcceptedCmp6']]
corr_Acc
```

```python
# mencari tahu apakah persebaran data merata atu tidak sehingga diapat metode yang tepat 
# untuk menguji korelasi antar variabel

from scipy.stats import kstest

for i in corr_Acc.columns:
    stats, pVal = kstest('norm',corr_Acc[i])
    print(i," : ", pVal)
```

```python
# Dari dugaan adanya korelasi tersebut, dapat dibuat hipotesis terhadap hal tersebut
H0 = 'Pada karakteristik yang dimiliki oleh customer dengan income yang kecil memiliki kecenderungan\nuntuk menolak atau menerima offer campaign lebih kecil atau sama dengan customer dengan income besar'
Ha = 'Pada karakteristik yang dimiliki oleh customer dengan income yang kecil memiliki kecenderungan\nuntuk menolak atau menerima offer campaign lebih besar dari customer dengan income besar'
```

```python
def hypot(h0,ha,pval):
    if pval < 0.05:
        print(f"H0 ditolak, maka {ha}")
    else:
        print(f"H0 diterima, maka {h0}")
```

```python
from scipy.stats import spearmanr
for col in corr_Acc.iloc[:,1:].columns:
    zstats, pval = spearmanr(corr_Acc['Income'], corr_Acc[col])
    print('Income,', col,' : ', pval)
print()
hypot(H0,Ha,pval)
```

**Analysis :**

Dengan menggunakan uji hipotesis didapat bahwa karakteristik customer dengan income yang kecil memiliki kecenderungan
untuk menolak atau menerima offer campaign lebih kecil atau sama dengan customer dengan income besar. Sehingga dapat dikatakan bahwa income tidak mempengaruhi kecenderungan customer untuk menolak atau menerima offer

```python
# Dari hasil uji normalitas, didapatkan data yang didapat merupakan non-parametric 
# sehingga metode yang tepat adalah spearman
corr_Acc = corr_Acc.corr(method='spearman')
corr_Acc
```
